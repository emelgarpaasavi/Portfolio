# Data Preprocessing and Creating a Model

```python
# Separate features from the label
# added .values  to return a Numpy array to avoid any issues with the pandas and numpy numerics
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Perform scaling after split to only fit the training set to avoid data leakage from the test set
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Create Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
# Base the number of neurons or units in our layers from the size of the actual feature data. Based on the shape,
# it looks like we have 19 incoming features.
model = Sequential()

# for optimum solution, let's use 5 layers 
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

# This has only 1 neuron since it will only be outputting the predicted price.
model.add(Dense(1))

# Compile model
model.compile(optimizer='adam',loss='mse') #since this is a regression model we'll use the 'mean squared error'
```

```python
# Train the model
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
# After each epoch of training the training data, validation quickly run the test data and check its loss
# We can then keep track of how it's performing not just on the training data, but also the test data.
# Keras only updates the weights and biases with training data, not test data.
# The smaller the batch size, the longer the training is going to take but it is likely to overfit the data since
# it does not pass the entire training data set at once. Instead, it only focuses on smaller batches.
```

# Model evaluation and Predictions

```python
losses = pd.DataFrame(model.history.history)
# This produces 2 sets of columns: 
# loss: loss on training
# val_loss: loss on test or validation
# This DataFrame is then used to check if we're overfitting the training data on the model.
```

```python
losses.plot()
# There's a decrease in both the training loss and the test/validation loss.
# There is no increase so far in the validation set towards the tail end so there is no improvement.
# After 400 epochs, it seems like the decrease happens much slower.
# If the loss went up after some of these epochs, it means that there is overfitting in the training data beccause
# now it has a much larger loss than the val_loss data (validation data)
```

```python
# evaluating test data
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)
```

```python
# Root mean squared error
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
df['price'].describe()
```

```python
5.402966e+05 # The average price of the house is around 540,000
```

```python
# Since the root mean squared error is around 30% of the mean price of the house, the result is not so great
# We could use the explained_variance_score to get a deeper meaning
# It tells how much variance is explained by the actual model
```

```python
explained_variance_score(y_test,predictions)
# this model performs okay. (1 is the best score)
```

```python
plt.figure(figsize=(12,5))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
# We have only explained around 80% of the variance in pricing.
# Also, we'll notice that we're basically punished by the outliers (expensive houses)
# Expensive houses are not good at predicting prices
```

```python
# What may be worth it is to retrain the model just on that bottom 99% of houses.
```

# Feature Engineering Process [Removed Extreme Values]

```python
# Change date dataframe to datetime, get months and years
non_top1_price['date'] = pd.to_datetime(non_top1_price['date'])
non_top1_price['month'] = non_top1_price['date'].dt.month
non_top1_price['year'] = non_top1_price['date'].dt.year
```

```python
# remove id, zipcode and date
non_top1_price = non_top1_price.drop(['id', 'date', 'zipcode'], axis=1)
```

# Data Preprocessing and Creating a Model [Removed Extreme Values]

```python
# split feature values and labels
X = non_top1_price.drop('price', axis=1).values
y = non_top1_price['price'].values
```

```python
# split training and test values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# perform standard scaling, create new instance of StandardScaler called 'non-extreme data' (ned) scaler
ned_scaler = StandardScaler()
# only fit the training set to avoid data leakage from the test set
X_train = ned_scaler.fit_transform(X_train)
X_test = ned_scaler.transform(X_test)
# get X_train shape
X_train.shape
```

```python
# create model
ned_model = Sequential()

ned_model.add(Dense(19, activation='relu'))
ned_model.add(Dense(19, activation='relu'))
ned_model.add(Dense(19, activation='relu'))

ned_model.add(Dense(1))

# compile model
ned_model.compile(optimizer='adam', loss='mse')
```

```python
ned_model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test), batch_size=128, epochs=400)
```

# Model Evaluation and Predictions [Removed Extreme Values]

```python
ned_losses = pd.DataFrame(ned_model.history.history)
```

```python
# plot non-extreme data losses
# as we can see in the plot, it has a similar trend with the previous dataframe with extreme values.
# there's a decrease in both training and validation/test losses
# as the epoch goes up to 300, there is not much of an improvement.
ned_losses.plot()
```

```python
# now, let's evaluate the test data
ned_predictions = ned_model.predict(X_test)
```

```python
# get the root mean squared error
np.sqrt(mean_squared_error(y_test, ned_predictions))
```

```python
# get mean of the non-extreme prices in the non_top1_price dataframe
non_top1_price['price'].mean()
```

```python
# compared with the mean of the dataframe with extreme prices, the mean of the non-extreme prices is quite lower
# mean with extreme prices : 540296.6
# mean with non-extreme prices : 518367.5
# the root mean squared error is around 27% of the mean, thus, this result also is not great
# let's get the explained variance score (how much variance is explained by the model) for clearer picture:
```

```python
explained_variance_score(y_test, ned_predictions)
# this model hovers around .8, which is okay
```

```python
# create a scatterplot of y_test and ned_predictions
plt.figure(figsize=(12,5))
plt.scatter(y_test,ned_predictions)
plt.plot(y_test,y_test,'r')
# as compared with the dataframe with extreme values, the scatterplot of the dataframe without extreme values does clump on 
# a specific range of prices so there are no extreme outliers found.
# we have also only explained around 80% of variance in pricing.
```
